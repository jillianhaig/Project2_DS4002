```python
# import necessary packages
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
# import atlantic ocean hurricane data
df = pd.read_csv('/content/Project2_DS4002/DATA/Individual Datasets/hurr.atl.days.csv')
```

```python
# rename column for clarity
df.rename(columns={'  Number of days with hurricane in Atlantic  missing value -99 https://psl.noaa.gov/data/timeseries/month/': 'hurricane_days'}, inplace=True)
```

```python
# change the date column to datetime format
df['date'] = pd.to_datetime(df['Date'])

# filter the data to start from 1970
df_filtered = df[df['date'] >= '1971-01-01']

# get the year from the date column and group by year and sum of hurricane days
df_filtered['year'] = df_filtered['date'].dt.year
yearly_data = df_filtered.groupby('year')['hurricane_days'].sum().reset_index()

# rename the columns for clarity
yearly_data.columns = ['Year', 'Hurricane_Days']
```

```python
# load second dataset for co2 emissions
df2 = pd.read_csv('/content/Project2_DS4002/DATA/Individual Datasets/annual-co2-emissions-per-country.csv')
```

```python
# filter for 1971 and later
df2 = df2[(df2['Entity'] == 'World') & (df2['Year'] >= 1971)]
```

```python
# drop extraneous column
df2.drop(columns=['Code'], inplace=True)
```

```python
# rename column for clarity
df2.rename(columns={'Annual COâ‚‚ emissions': 'CO2_emissions'}, inplace=True)
```

```python
# merge first 2 datasets
merged_df = pd.merge(yearly_data, df2, on='Year')
```

```python
# drop column (not needed)
merged_df = merged_df.drop(columns=['Entity'])
```

```python
# rename columns
merged_df.rename(columns={'Hurricane_Days': 'Atlantic_Hurricane_Days'}, inplace=True)
```

```python
merged_df
```

```python
# load next dataset for pacific ocean hurricane data
df3 = pd.read_csv('/content/Project2_DS4002/DATA/Individual Datasets/hurr.nepac.days.csv')
```

```python
# find column names
column_names = df3.columns.tolist()

print(column_names)
```

```python
# rename column for clarity
df3.rename(columns={'  NE Pacific Hurricane days/month  missing value -99 https://psl.noaa.gov/data/timeseries/month/': 'Pacific_Hurricane_Days'}, inplace=True)
```

```python
# change year data to date format
df3['Date'] = pd.to_datetime(df3['Date'])

# filter the data to only include data from 1971 onwards
df_filtered2 = df3[df3['Date'] >= '1971-01-01']

# get the year from the date
df_filtered2['Year'] = df_filtered2['Date'].dt.year

# group by year and get the sum
yearly_avg2 = df_filtered2.groupby('Year')['Pacific_Hurricane_Days'].sum().reset_index()

# rename the columns for clarity
yearly_avg2.columns = ['Year', 'Pacific_Hurricane_Days']
df_3 = yearly_avg2
```

```python
# merge with previous datasets
merged_df2 = pd.merge(merged_df, df_3, on='Year')
```

```python
# add new column with the sum of pacific and atlantic hurricane days
merged_df2['Total_Hurricane_Days'] = merged_df2['Pacific_Hurricane_Days'] + merged_df2['Atlantic_Hurricane_Days']
```

```python
# load new dataset for population
df_pop = pd.read_csv('/content/Project2_DS4002/DATA/Individual Datasets/population (1).csv')
```

```python
# rename and drop columns for clarity
df_pop.rename(columns={'Population (historical)': 'Population'}, inplace=True)
df_pop.drop(columns=['Code'], inplace=True)
```

```python
# filter for years and drop column
population_df = df_pop[(df_pop['Entity'] == 'World') & (df_pop['Year'] >= 1971) & (df_pop['Year'] <= 2021)]
population_df.drop(columns=['Entity'], inplace=True)
```

```python
# merge data
merged_df3 = pd.merge(merged_df2, population_df, on='Year')
```

```python
# load in sea level dataset
sealevel_df = pd.read_csv('/content/Project2_DS4002/DATA/Individual Datasets/sea-level_fig-1.csv')
```

```python
# rename column for clarity
sealevel_df.rename(columns={'Figure 1. Global Average Absolute Sea Level Change, 1880-2023': 'Year'}, inplace=True)
```

```python
# drop first 5 rows, these are just info about the dataset (not actual data we want to include)
dfsea_dropped = sealevel_df.drop(sealevel_df.index[:5])
```

```python
# show column names
column_names = dfsea_dropped.columns.tolist()
print(column_names)
```

```python
# drop extraneous columns and first row
dfsea_dropped.rename(columns={'Unnamed: 1': 'CSIRO_Adjusted_SeaLevel_Inch'}, inplace=True)
dfsea_dropped.drop(columns=['Unnamed: 2'], inplace=True)
dfsea_dropped.drop(columns=['Unnamed: 3'], inplace=True)
dfsea_dropped.drop(columns=['Unnamed: 4'], inplace=True)
dfsea_dropped = dfsea_dropped.drop(dfsea_dropped.index[:1])
```

```python
# change the string type year data to integers
dfsea_dropped['Year'] = dfsea_dropped['Year'].astype(int)
```

```python
# filter the years we want
sealevel_filtered = dfsea_dropped[(dfsea_dropped['Year'] >= 1971) & (dfsea_dropped['Year'] <= 2021)]
```

```python
# merge data
merged_df4 = pd.merge(merged_df3, sealevel_filtered, on='Year')
```

```python
# load in data for earth surface temp change
surface_df = pd.read_csv('/content/Project2_DS4002/DATA/Individual Datasets/Annual_Surface_Temperature_Change.csv')
```

```python
# get the column names
column_names = surface_df.columns.tolist()
print(column_names)
```

```python
# drop unneeded columns
surface_df.drop(columns=['ObjectId'],inplace=True)
surface_df.drop(columns=['ISO2'],inplace=True)
surface_df.drop(columns=['ISO3'],inplace=True)
surface_df.drop(columns=['Indicator'],inplace=True)
surface_df.drop(columns=['Unit'],inplace=True)
surface_df.drop(columns=['Source'],inplace=True)
surface_df.drop(columns=['CTS_Code'],inplace=True)
surface_df.drop(columns=['CTS_Name'],inplace=True)
surface_df.drop(columns=['CTS_Full_Descriptor'],inplace=True)
```

```python
# filter just the world temperature
surface_filtered = surface_df[surface_df['Country'] == 'World']
```

```python
# pivot the data and change the year variable to be the correct format and type
surface_filtered.columns = surface_filtered.columns.str.replace('F', '', regex=False)
melted_df = surface_filtered.melt(id_vars=['Country'], var_name='Year', value_name='Surface Temperature')

# filter for years between 1971 and 2021
filtered_sdf = melted_df[(melted_df['Year'].astype(int) >= 1971) & (melted_df['Year'].astype(int) <= 2021)]

# drop the country column (we only need year and surface temperature)
filtered_sdf = filtered_sdf.drop(columns=['Country'])
```

```python
# make year into an int
filtered_sdf['Year'] = filtered_sdf['Year'].astype(int)
```

```python
# rename column
filtered_sdf.rename(columns={'Surface Temperature': 'Surface_Temperature_Change'}, inplace=True)
```

```python
# merge data
merged_df5 = pd.merge(merged_df4, filtered_sdf, on='Year')
```

```python
# import next dataset for sea temp change
seatemp_df = pd.read_csv('/content/Project2_DS4002/DATA/Individual Datasets/sea-surface-temp_fig-1.csv', encoding='ISO-8859-1')
```

```python
# get the column names
column_names = seatemp_df.columns.tolist()
print(column_names)
```

```python
# rename column
seatemp_df.rename(columns={'Figure 1. Average Global Sea Surface Temperature, 1880-2023': 'Year'}, inplace=True)
```

```python
# drop into added to top of dataset
seatemp_dropped = seatemp_df.drop(seatemp_df.index[:5])
```

```python
# drop unneeded columns
seatemp_dropped.drop(columns=['Unnamed: 2'], inplace=True)
seatemp_dropped.drop(columns=['Unnamed: 3'], inplace=True)
seatemp_dropped.rename(columns={'Unnamed: 1': 'SurfaceSeaTemp_Annual_Anomaly'}, inplace=True)
```

```python
# drop top column
seatemp_dropped = seatemp_dropped.drop(seatemp_dropped.index[:1])
```

```python
# make year into an int
seatemp_dropped['Year'] = seatemp_dropped['Year'].astype(int)
```

```python
# filter for years we want
seatemp_data = seatemp_dropped[(seatemp_dropped['Year'] >= 1971) & (seatemp_dropped['Year'] <= 2021)]
```

```python
# merge data
hurricane_data = pd.merge(merged_df5, seatemp_data, on='Year')
```

```python
# swap order of the columns for clarity
cols = hurricane_data.columns.tolist()
cols[2], cols[3] = cols[3], cols[2]
hurricane_data = hurricane_data[cols]
```

```python
# swap order of the columns for clarity
cols2 = hurricane_data.columns.tolist()
cols2[3], cols2[4] = cols2[4], cols2[3]
hurricane_data = hurricane_data[cols2]
```
