```python
# import packages
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
# load in final hurricane dataset
hurricane_data = pd.read_csv('/content/Project2_DS4002/DATA/Final Dataset/Hurricane_Dataset.csv')
```

```python
# atlantic trend
sns.regplot(data=hurricane_data, x='Year', y='Atlantic_Hurricane_Days',
            scatter_kws={'color': 'blue', 's': 25},
            line_kws={'color': 'black'})
```

```python
# pacific trend
sns.regplot(data=hurricane_data, x='Year', y='Pacific_Hurricane_Days',
            scatter_kws={'color': 'green', 's': 25},
            line_kws={'color': 'black'})
```

```python
# atlantic + pacific trend
sns.regplot(data=hurricane_data, x='Year', y='Total_Hurricane_Days',
            scatter_kws={'color': 'purple', 's': 25},
            line_kws={'color': 'black'})
```

```python
# line graph for total hurricane days
sns.set(style='whitegrid')
plt.figure(figsize=(10, 6))
sns.lineplot(data=hurricane_data, x='Year', y='Total_Hurricane_Days', marker='o')

# titles and labels
plt.title('Total Hurricane Days 1971-2021 (Atlantic and Pacific Oceans)', fontsize=16)
plt.xlabel('Year', fontsize=14)
plt.ylabel('Total Hurricane Days', fontsize=14)

# show the plot
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# stacked area plot
hurricane_data.set_index('Year', inplace=True)  # set year as the index

plt.figure(figsize=(10, 6))
plt.stackplot(hurricane_data.index, hurricane_data['Pacific_Hurricane_Days'], hurricane_data['Atlantic_Hurricane_Days'], hurricane_data['Total_Hurricane_Days'],
               labels=['Pacific Hurricane Days', 'Atlantic Hurricane Days', 'Total Hurricane Days'])

# titles and labels
plt.title('Stacked Area Plot of Hurricane Days', fontsize=16)
plt.xlabel('Year', fontsize=14)
plt.ylabel('Number of Hurricane Days', fontsize=14)
plt.legend(loc='upper left')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# plot relationship between CO2 emissions and total hurricane days
sns.regplot(data=hurricane_data, x='Total_Hurricane_Days', y='CO2_emissions',
            scatter_kws={'color': 'black', 's': 25},
            line_kws={'color': 'black'})
```
